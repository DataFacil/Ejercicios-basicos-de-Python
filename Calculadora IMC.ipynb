Construya una función que reciba como argumentos el peso y altura de una persona, y devuelva el IMC e indique la clasificación, de acuerdo a la siguiente tabla:

Y la siguiente fórmula:

```python
def IMC(peso, estatura):
    imc = peso / estatura**2
    if imc <= 18.4:
        print(f'El IMC es {round(imc, 2)}, lo que se clasifica como delgadez')
    elif imc > 18.4 and imc <= 24.9:
        print(f'El IMC es {round(imc, 2)}, lo que se clasifica como peso saludable')
    elif imc > 24.9 and imc <= 29.9:
        print(f'El IMC es {round(imc, 2)}, lo que se clasifica como sobrepeso')
    else:
        print(f'El IMC es {round(imc, 2)}, lo que se clasifica como obesidad')
```
